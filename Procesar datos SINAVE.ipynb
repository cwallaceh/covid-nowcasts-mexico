```python
import pandas as pd
import json
```

```python
with open('estados_sa.txt') as f:
  edata = json.load(f)

# Output: {'name': 'Bob', 'languages': ['English', 'Fench']}
dfr = pd.read_json(edata["d"])
dfr.rename(columns={0:"Estado",1:"Region"},inplace=True)
dfr.set_index("Region",inplace=True)
dfr
```

```python
with open('datos_1.txt') as f:
  data = json.load(f)

df = pd.read_json(data["d"])
df.rename(columns={0:"Region",3:"Acumulados",6:"Tasa de incidencia"},inplace=True)

df["Fecha"] = pd.to_datetime(df[5],dayfirst=True)

df.drop(columns={5}, inplace=True)

dfa = df[["Region","Fecha","Acumulados"]]

dfa.set_index("Region",inplace=True)

dfa = dfa.join(dfr, how="left")

dfa = dfa.reset_index()

# Ordenar por indice
dfa = dfa.set_index(["Fecha","Region"]).sort_index(level=("Region","Fecha"))

#Shiftear un dia hacia adelante
dfs = dfa.shift(1)
dfa["Anterior"] = dfs["Acumulados"]

dfa["Casos"] = dfa["Acumulados"] - dfa["Anterior"]
dfa
```

```python
with open('datos_2.txt') as f:
  data = json.load(f)

df = pd.read_json(data["d"])
df.rename(columns={0:"Region",3:"Acumulados",6:"Tasa de incidencia"},inplace=True)

df["Fecha"] = pd.to_datetime(df[5],dayfirst=True)

df.drop(columns={5}, inplace=True)

dfa2 = df[["Region","Fecha","Acumulados"]]

dfa2.set_index("Region",inplace=True)

dfa2 = dfa2.join(dfr, how="left")

dfa2 = dfa2.reset_index()

# Ordenar por indice
dfa2 = dfa2.set_index(["Fecha","Region"]).sort_index(level=("Region","Fecha"))

#Shiftear un dia hacia adelante
dfs = dfa2.shift(1)
dfa2["Anterior"] = dfs["Acumulados"]

dfa2["Casos"] = dfa2["Acumulados"] - dfa2["Anterior"]
dfa2
```

```python
dfa = dfa.append(dfa2)
```

```python
casos = pd.DataFrame()
casos = dfa[["Estado","Casos"]]
casos.reset_index(inplace=True)
casos.dropna(inplace=True)

casos.rename(columns={"Region":"Region_ID","Estado":"region","Casos":"cases","Fecha":"date"}, inplace=True)

casos.drop(columns={"Region_ID"},inplace=True)

casos["import_status"] = "local"

#casos[casos["region"] != "NACIONAL"]

casos.to_csv("casos_270520.txt", index=False)
```
